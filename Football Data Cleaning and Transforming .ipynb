```python
def money(x):
    if "€" in x:
        x = x.replace("€", "")

    if "M" in x:
        x = x.replace("M", "")
        return int(float(x) * 1_000_000)

    elif "K" in x:
        x = x.replace("K", "")
        return int(float(x) * 1_000)

    return int(x)
```

```python
df["Value"] = df["Value"].apply(money) / 1_000_000
df["Wage"] = df["Wage"].apply(money)
df["Release Clause"] = df["Release Clause"].apply(money) / 1_000_000

df=df.rename(columns={"Value": "Values in Euro Million","Wage": "Wage in Euros","Release Clause": "Release Clause in Euro Million"})
```

```python
from datetime import datetime
```

```python
def type(x):
    if 'Free' in x:
        return "Free"
    if "Loan" in x:
        return "Loan"
    if "~" in x:
        return "Contract"
    else:
        return pd.na
```

```python
def start_time_contract(x):
    if '~' in x:
        return int(x[:4])
    if "Loan" in x:
        # Replace both lowercase and uppercase variations if needed
        x = x.replace(" on Loan", "")
        x = x.replace(" On Loan", "")
        x = x.strip()
        # Now parse the date
        x = datetime.strptime(x, "%b %d, %Y")
        return x.date()
    else:
        return pd.NA
    
def end_time_contract(type,contract,loan):
    if type == "Contract":
        return int(contract[-4:])
    if type == "Loan":
        loan = datetime.strptime(loan, "%b %d, %Y")
        return loan.date()
    else:
        return pd.NA
```

```python
df["Type of contract"] = df["Contract"].apply(type)
```

```python
df["End year"] = df.apply(lambda row: end_time_contract(row["Type of contract"],row["Contract"],row["Loan Date End"] ), axis=1)
```

```python
df["Start year"] = df["Contract"].apply(start_time_contract)
```
